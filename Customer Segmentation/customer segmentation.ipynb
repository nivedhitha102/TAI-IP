```python
# Feature Engineering

#Age of customer today 
df['Age']=2023-df['Year_Birth']


#Total spendings
df['expenses']= df["MntWines"]+ \
                df["MntFruits"]+ df["MntMeatProducts"]+ \
                df["MntFishProducts"]+ df["MntSweetProducts"]+ \
                df["MntGoldProds"]

# marital status
df['Marital_Status']=df['Marital_Status'].replace({'Married':'couple',
                                                 'Together':'couple',
                                                 'Divorced':'Single',
                                                 'Widow':'Single',
                                                 'Alone':'Single',
                                                 'Absurd':'Single',
                                                 'YOLO':'Single'})

# total children living in the family
df['children']=df['Kidhome']+df['Teenhome']

# total members in the family
df['family_count']=df['Marital_Status'].replace({'couple':2,'Single':1})\
                   +df['children']
```

# DATA PREPROCESSING

```python
# Scaling the features using the standard scaler
sc=StandardScaler()
df_scaled=pd.DataFrame(sc.fit_transform(df),columns=df.columns)
df_scaled
```

# DIMENSIONALITY REDUCTION

```python
# Principal component analysis (PCA) 
# checking for the best optimal value by finding the eigen values geater than 1

cov_matrix=np.cov(np.transpose(df_scaled),ddof=1)
cov_matrix
```

```python
eig_val,eig_vector=np.linalg.eig(cov_matrix)
print('eig_val==',eig_val,'\n','eig_vectors==',eig_vector)
```

```python
(eig_val>1).sum()
#optimal number is 5
```

```python
eig_value=eig_val.tolist()
eig_value.sort(reverse=True)
```

```python
exv=[]
for i in eig_value:
    exv.append((i/sum(eig_value))*100)
exv
```

```python
np.cumsum(exv)
```

```python
#scree plot 

plt.plot(eig_value,marker='*')
plt.xlabel('pcs')
plt.ylabel('eigval')
plt.axvline(5)  # optimal number (5)
plt.show()
```

```python
#Initiating PCA to reduce dimentions to 5
pca = PCA(n_components=5)
pca.fit(df_scaled)
PCA_df = pd.DataFrame(pca.transform(df_scaled), columns=(["PC0","PC1", "PC2",'PC3','PC4']))
PCA_df.describe().T
```

# Agglomerative clustering

```python
euci_dist = euclidean_distances(df_scaled)
euci_dist
```

```python
l=linkage(df_scaled)
l
```

```python
plt.figure(figsize=(20,20))
l=linkage(df_scaled,method='ward')
dendrogram(l)
plt.show()

# It shows that the best clusters from dendrogram is 2
```

```python
e=euci_dist[np.triu_indices(2240,k=1)]
```

```python
cophenet(l,e)
```

```python
#Initiating the Agglomerative Clustering model 
AC = AgglomerativeClustering(n_clusters=2)

AC.fit_predict(PCA_df)
PCA_df["labels"] = AC.labels_
```

```python
df['labels']=AC.labels_
```
